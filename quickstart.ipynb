## Classical racing

This quickstart is designed to show you the process of using the framework to run tests.

1. Generate racelines
    - You can inspect the raceline images in the `Data/raceline_data/` folder
2. Run test laps with the pure pursuit planner
3. Plot data
    - You can inspect the images in the `Logs/GlobalPlanPP/mu70` folder


> [!note] Configuration
> All the parameters for the simulator and planners are in the `params/` directory

```python
from f1tenth_sim.classic_racing.RaceTrackGenerator import RaceTrackGenerator, RaceTrackPlotter

params = load_parameter_file("RaceTrackGenerator")
params.mu = 0.7
raceline_id = f"mu{int(params.mu*100)}"
map_list = ['aut', 'esp', 'gbr', 'mco']
for map_name in map_list: 
    RaceTrackGenerator(map_name, raceline_id, params)
    RaceTrackPlotter(map_name, raceline_id)
```

```python
from f1tenth_sim.classic_racing.GlobalPurePursuit import GlobalPurePursuit


test_id = "mu70"
map_name = "aut"
planner = GlobalPurePursuit(test_id, False, planner_name="GlobalPlanPP")
test_planning_all_maps(planner, test_id)

```

```python
from f1tenth_sim.data_tools.general_plotting.plot_trajectory_analysis import plot_trajectory_analysis
from f1tenth_sim.data_tools.general_plotting.plot_raceline_tracking import plot_raceline_tracking

plot_trajectory_analysis(planner.name, test_id)
plot_raceline_tracking(planner.name, test_id)
```
